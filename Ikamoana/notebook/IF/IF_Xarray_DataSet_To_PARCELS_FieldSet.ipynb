```python
import ikamoana
import parcels
```

```python
fh = ikamoana.feedinghabitat.FeedingHabitat(
    xml_filepath="./../../data/feeding_habitat_data/po_2x30_historical_interim/skj_po_interim_2deg_configuration_file.xml")
```

## Creating DataSet using **Ikamoana Feeding Habitat** module

```python
fh_cohort_0_to_3 = fh.computeFeedingHabitat([0,1,2,3])
fh_cohort_0_to_3
```

## Convert **Xarray DataSet** into **PARCELS FieldSet**

```python
variables = {
    'C0': 'Feeding_Habitat_Cohort_0',
    'C1': 'Feeding_Habitat_Cohort_1',
    'C2': 'Feeding_Habitat_Cohort_2',
    'C3': 'Feeding_Habitat_Cohort_3'}

dimensions = {
    'lon': 'lon',
    'lat': 'lat',
    'time': 'time'}
```

>**Error** on latitude axis that is ordered **North-South** and should be ordered **South-Nord**

```python
parcels.FieldSet.from_xarray_dataset(fh_cohort_0_to_3, variables=variables, dimensions=dimensions)
```


## **Flip** latitude axis and **reindex** DataSet

```python
fh_cohort_0_to_3.reindex(lat=list(reversed(fh_cohort_0_to_3.lat)))
```

```python
parcels.FieldSet.from_xarray_dataset(
    fh_cohort_0_to_3.reindex(lat=list(reversed(fh_cohort_0_to_3.lat))),
    variables=variables,
    dimensions=dimensions)
```
