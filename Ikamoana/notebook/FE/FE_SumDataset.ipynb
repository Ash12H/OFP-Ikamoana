## Sum of Effort (or catch)

```python
filepath = "./../../data/fisheries_effort/skj_fisheries_data_2021_Nov17.txt"
```

```python
fisheries_df = fe.readFile(filepath, header_position=0, columns_name=None)
fisheries_dict = fe.separateFisheries(fisheries_df)
fisheries_rescaled_dict = fe.rescaleFisheries({1:fisheries_dict[1],15:fisheries_dict[15]})
fisheries_groub_by_dict = fe.groupByFisheries(fisheries_rescaled_dict)
fisheries_effort_ds, fisheries_catch_ds = fe.fisheriesToDataSet(fisheries_groub_by_dict)
fisheries_effort_ds
```

```python
effort_sum = fe.sumDataSet(fisheries_effort_ds)
effort_sum
```

```python
print("Original : ",float(fisheries_effort_ds[1].sum())+float(fisheries_effort_ds[15].sum()))
print("Sum      : ",float(effort_sum.sum()))
```

> ### Verification of the coords size (continuous ?)

```python
lat_o = effort_sum.lat.data
lat_s = np.arange(effort_sum.lat.data[0],effort_sum.lat.data[-1]+0.001, step=1.0)

np.sum(lat_o == lat_s) == lat_s.size == lat_o.size
```

```python
lon_o = effort_sum.lon.data
lon_s = np.arange(effort_sum.lon.data[0],effort_sum.lon.data[-1]+0.001, step=1.0)

np.sum(lon_o == lon_s) == lon_s.size == lon_o.size
```

```python
time_o = np.array(effort_sum.time.data,dtype='datetime64[M]')
time_s = np.arange(time_o[0],time_o[-1]+1)

np.sum(time_o == time_s) == time_s.size == time_o.size
```
