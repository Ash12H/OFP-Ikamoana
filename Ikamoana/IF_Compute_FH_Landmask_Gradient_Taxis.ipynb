```python
import ikamoana
```

## Initialize

```python
xml_filepath="./data/feeding_habitat_data/po_2x30_historical_interim/skj_po_interim_2deg_configuration_file.xml"

fh = ikamoana.feedinghabitat.FeedingHabitat(xml_filepath)
ikaf = ikamoana.ikamoanafields.IkamoanaFields(xml_filepath,xml_filepath)
```

## 1. Compute Taxis for Cohort number 0
### 1.1 Compute Feeding Habitat

```python
data = fh.computeFeedingHabitat(0)['Feeding_Habitat_Cohort_0']
data
```

### 1.2 Compute Gradient

```python
grad_data_lon, grad_data_lat = ikaf.gradient(data,ikaf.landmask(data))
grad_data_lon
```

### 1.3 Compute Taxis

```python
taxis_data_lon, taxis_data_lat = ikaf.taxis(grad_data_lon, grad_data_lat)
taxis_data_lon
```

## 2. Compute Taxis for a Cohort that evolves
### 2.1 Compute Feeding Habitat

```python
data_evolving = fh.computeEvolvingFeedingHabitat()
data_evolving
```

### 2.2 Compute Gradient

```python
grad_evolve_lon, grad_evolve_lat = ikaf.gradient(data_evolving, ikaf.landmask(data_evolving))
grad_evolve_lon
```

### 2.3 Compute Taxis

```python
taxis_evolve_lon, taxis_evolve_lat = ikaf.taxis(grad_evolve_lon, grad_evolve_lat)
taxis_evolve_lon
```
