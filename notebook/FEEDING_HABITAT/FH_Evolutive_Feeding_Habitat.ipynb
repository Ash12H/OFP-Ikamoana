# Select sub-part using the cohorts dimension instead of time dimension

```python
tmp_1 = tmp.swap_dims({"time":"cohorts"})
print("First, we swapt the dimension time to cohort :\n",tmp.coords)
tmp_1 = tmp_1.sel(cohorts=36)
print("\n\nThen we can select a partition of the DataArray.\n"
      +"For example, we select here the oldest cohort :\n\n")
tmp_1
```

Verification with the `computeFeedingHabitat` method.

```python
tmp_2 = fh.computeFeedingHabitat(cohorts=36,time_start=366,time_end=365+18)

tmp_1 = tmp_1.data
tmp_2 = tmp_2.to_array()[0,:,:,:].data
```

```python
print("tmp_1 shape : ",tmp_1.shape)
print("tmp_2 shape : ",tmp_2.shape)
print("Mean (Substraction) : ",np.nanmean(tmp_1 - tmp_2))
print("Sum (Absolute Substraction): ",np.nansum(np.absolute(tmp_1 - tmp_2)))
```
